```python
#Dropping Rows with Missing Values

data.dropna(inplace=True)
```

```python
#Checking for Missing Values Again
data.isnull().sum()
```

```python
# Checking for Duplicate Rows
#Removing Duplicate Rows
#Rechecking for Duplicates

data.duplicated().sum()
data.drop_duplicates(inplace=True)
data.duplicated().sum()
```

```python
#Checking Data Types
data.dtypes
#data['id'] = data['id'].astype(object)
data['id'] = data['id'].astype(object)
#Checking Data Types Again
data.dtypes
# Changing host_id Column to object
data['host_id'] = data['host_id'].astype(object)
#Checking Data Types One More Time
data.dtypes
```

```python
#Filtering the Data
df = data[data['price'] < 1500]

#Creating a Boxplot
sns.boxplot(data=df, x='price')
```

```python
#Set the Figure Size
plt.figure(figsize=(8, 5))

#Create a Histogram with Seaborn
sns.histplot(data=df, x='price', bins=100)

#Add a Title
plt.title('Price Distribution')

#Add Y-axis Label
plt.ylabel("Frequency")

#Show the Plot
plt.show()
```

```python
df.dtypes
```

```python
#Groups the DataFrame df by the column neighbourhood_group.Each unique neighbourhood group becomes one group.
#For each group, calculates the mean (average) of the price column.

df.groupby(by='neighbourhood_group')['price'].mean()
```

Feature Engineering


```python
# Creating a safe, independent copy of the filtered dataset
df = data[data['price'] < 1500].copy()   # .copy() avoids SettingWithCopyWarning

# Calculating price per bed for each listing
df['price per bed'] = df['price'] / df['beds']

# Grouping by neighbourhood group and calculating average price per bed
df.groupby(by='neighbourhood_group')['price per bed'].mean()
```

Bi Variable Analysis One variable depenency in another variable

```python
df.columns
```

```python
# price dependency on neighbourhood
sns.barplot(data=df, x='neighbourhood_group', y='price', hue='room_type')
```

```python
# number of reviews and price rel
plt.figure(figsize=(8, 5))  # Set figure size
plt.title("Locality and Review Dependency")  # Set plot title

# Scatter plot showing relationship between number of reviews and price
sns.scatterplot(data=df, x='number_of_reviews', y='price', hue='neighbourhood_group')

plt.show()  # Display the plot
```

```python
df.dtypes
```

```python
# Visualizing pairwise relationships between key variables
# Includes price, minimum nights, number of reviews, and availability
# Colored by room_type to compare patterns across different room types
sns.pairplot(data=df, 
             vars=['price', 'minimum_nights', 'number_of_reviews', 'availability_365'], 
             hue='room_type')
```

```python
# Scatter plot showing geographical distribution of listings based on longitude and latitude
# Colored by room_type to observe how different room types are spread across the map
plt.figure(figsize=(10, 7))  
sns.scatterplot(data=df, x='longitude', y='latitude', hue='room_type')
plt.title("Geographical Distribution of Airbnb Listings")
plt.show()
```

```python
df.dtypes
```

```python
# Heatmap to visualize correlation between numerical variables
# Helps identify how strongly numerical columns relate to each other
# Correlation ranges from -1 (perfect negative) to +1 (perfect positive)

# Calculating correlation matrix for selected numerical columns
corr = df[['latitude', 'longitude', 'price', 'minimum_nights', 'number_of_reviews', 
           'reviews_per_month', 'availability_365', 'beds']].corr()

# Display correlation matrix
corr

# Plotting the heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(data=corr, annot=True, cmap='coolwarm')  # annot=True shows correlation values
plt.title("Correlation Heatmap of Numerical Features")
plt.show()
```
